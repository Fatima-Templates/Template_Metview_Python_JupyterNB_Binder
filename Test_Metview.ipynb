```python
import metview as mv
```

```python
# Running test commands

test = mv.read("2t.grib") # read fieldset

vals = mv.values(test) # Convert fieldset in vector
lats = mv.latitudes(test)
lons = mv.longitudes(test)
vals = vals[1:100]
lats = lats[1:100]
lons = lons[1:100]
print(mv.type(vals))

geo = mv.create_geo(100) # create geopint
geo = mv.set_latitudes(geo, lats)
geo = mv.set_longitudes(geo, lons)
geo = mv.set_values(geo, vals)
```

```python
# To display plots within the Jupyter notebook
mv.setoutput("jupyter")
```

```python
# Plot
mv.plot(test)
```
